```python
# Importing pandas library
import pandas as pd

# Reading and displaying csv file
df = pd.read_csv("DataScienceSalaries_8b290669-a5e9-45bf-be72-d27add2eacae_93472_.csv")
df
```

```python
# Checking all information regarding the variables
df.info()
```

```python
# Dropping two columns without inplace = True
df.drop(columns = ['salary','salary_currency'])
df
```

```python
# Dropping two columns
df.drop(columns = ['salary','salary_currency'], inplace = True)
df
```

```python
# Checking for NaN values
df.isna()
```

```python
# Method to drop NaN values
df.dropna()
```

```python
# Importing matplotlib
import matplotlib.pyplot as plt

# Grouping job title by each unique value, and counting them
top15 = df.groupby('job_title').size()

# Then finding the largest 15 values from them
top15_jobs = top15.nlargest(15)
top15_jobs
```

```python
# Plotting the above data
plt.bar(top15_jobs.index, top15_jobs.values, ec ='black', color = 'powderblue', width = 1)
plt.title("Top 15 jobs")
plt.xlabel("Job Titles")
plt.ylabel("Occurrences")
plt.xticks(rotation = 90)

# Display the plot
plt.show()
```

```python
# Importing matplotlib
import matplotlib.pyplot as plt

# Calculating the top 5 highest salaries with nlargest function
top5 = df.nlargest(5, 'salary_in_usd')[['job_title', 'salary_in_usd']]
top5
```

```python
# Plotting the above data
cname = ['rosybrown','lightpink','mistyrose','lavenderblush','snow'] # variable storing colors
plt.bar(top5['job_title'], top5['salary_in_usd'], color=cname, width = 0.9, ec = 'pink')
plt.xlabel('Job Titles')
plt.ylabel('Salaries in US dollars')
plt.title('Top 5 Jobs with the Highest Salaries')
plt.xticks(rotation=90)

# Display the plot
plt.show()
```

```python
# Importing matplotlib
import matplotlib.pyplot as plt

# Calculating the average salary with mean for each experience level
avg_salary = df.groupby('experience_level')['salary_in_usd'].mean()

# Find the experience levels with the highest average salaries
experiencelevels = avg_salary.nlargest()
experiencelevels

```

```python
# Now to plot this data
plt.bar(experiencelevels.index, experiencelevels.values, color = 'lavender', ec = 'brown', width = 0.9)
plt.title('Average salary per Experience level')
plt.xlabel('Experience levels')
plt.ylabel('Average salary')
plt.xticks(rotation = 90)

# Display the plot
plt.show()
```

```python
# Importing matplotlib
import matplotlib.pyplot as plt

# Creating histogram
plt.hist(df['salary_in_usd'], bins=12, ec = 'k', linewidth = 0.7, color='azure')
plt.title('Histogram of ' + 'salary_in_usd')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating box plot
plt.boxplot(df['work_year'])
plt.title('Box plot of ' + 'work_year')
plt.ylabel('work years')
plt.xlabel('all workers')

# Showing plot
plt.show()
```
